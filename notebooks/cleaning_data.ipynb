```python
import pandas as pd
```

```python
# Read data 
df = pd.read_csv('../data/employee_data.csv')
df.info()
```

```python
df.describe(include='all')
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

## Drop rows where more than two columns are missing data.

```python
# Drops rows with more than 2 missing values
df = df.dropna(thresh=len(df.columns) - 2)
df
```

## Fill missing values in the Salary column with the average salary.

```python
# Replace N/A sales values with their mean
df = df.fillna(df['Salary'].mean())
print(df)
```

```python
df.to_csv('cleaned_employee_data.csv')
```
